```python
people=['ann','brandan','chen','kavya','shivani','neha',]
age=[12,13,14,15,16,17]
weight=[55,45,67,45,33,44]
height=[160,216,170,130,120,140]
```

```python
import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("relationship between weight and height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight in kilos")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')
#main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='pink',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='red',edgecolor='k',alpha=0.75,bins=5)
plt.title("histogram of patitent weight",fontsize=18)
plt.xlabel("weight in kgs",fontsize=15)
plt.show()
```
